```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
template = np.zeros((200,200))
center = (100,100)
dist_horizontal = center[1] ## Column num is horizontal 
dist_vertical = center[0] ## Row num is vertical
for i in range(200):
    for j in range(200):
        template[i,j] = np.sqrt((dist_vertical-i)**2 + (dist_horizontal-j)**2)
plt.subplot(1,3,1)        
plt.imshow(template)
plt.subplot(1,3,2)
plt.imshow(template < 100, 'gray')

bigcircle = np.zeros((200,200))
template < 100 = 0
```

```python
template = np.zeros((200,200))
center = (100,100)
dist_horizontal = center[1] ## Column num is horizontal 
dist_vertical = center[0] ## Row num is vertical
for i in range(200):
    for j in range(200):
        template[i,j] = np.sqrt((dist_vertical-i)**2 + (dist_horizontal-j)**2)
        
plt.subplot(1,3,1)
plt.imshow(template < 100, 'gray')
bigcircle = template

for i in range(200):
    for j in range(200):
        if bigcircle[i][j] < 50 or bigcircle[i][j] > 100:
            bigcircle[i][j] = 0

plt.subplot(1,3,2)
plt.imshow(bigcircle)
```

```python
tmpl = np.zeros((200,200))
slope = 140 / 120
zeropoint = 140
for row in range(200):
    for col in range(200):
        if row + (slope * col) < zeropoint:
            tmpl[row][col] = 255
plt.imshow(tmpl)
```

```python
tmpl = np.zeros((200,200))
l1slope = 140 / 120
l1point = 140
l2slope = -(150 / 79)
l2point = 120*(l2slope)
l3slope = -(10/199)
l3point = 140

for row in range(200): ### row is like y in function
    for col in range(200):  ### col is like x in function
        if row + l1slope * col < l1point:
            tmpl[row][col] = 255
        if row + l2slope * col < l2point:
            tmpl[row][col] = 220
        if row + l3slope * col > l3point:
            tmpl[row][col] = 150
plt.imshow(tmpl, 'gray')
```

```python
black = np.zeros((50,50))
white = np.full((20,20), 255)
white[0:20,0:20] = black
plt.imshow(white, "gray")
```
